**Methods and Plan**

To answer the question
- plan to use k-nearest neighbors classification
- need to assume that the main idea behind it is that players who are close to each other based on experience and total hours played are likely to behave in similar ways as in if they would be subscribed or not
- However there are a few downsides to k-NN.
    - method depends heavily on how variables are scaled
        - will need to standardize played_hours
    - imbalance in hours played and subscription levels can affect the results of our k-NN classification
    - need to prepare the experience variable for it to be included in the model
- then use cross validation to acquire the best k value
    - the data will first be split into a training set and a test set using prop = 0.75 which will give a 75/25 split
- on the training set,  run 5-fold cross-validation to compare different k values
    - then select the one with the lowest classification error which will be used for the testing data

Before doing classification:
- tidy the data
    - converting experience into a factor
    - standardizing the numeric variables
    - removing missing values such as in Age

```r

players_split <- initial_split(players, prop = 0.75, strata = subscribe)
players_train <- training(players_split)
players_test <- testing(players_split)


player_vfold <- vfold_cv(players_train, v = 5, strata = subscribe)
player_recipe <- recipe(subscribe ~ played_hours + experience, data = players_train) |>
                    step_scale(all_numeric_predictors()) |>
                    step_center(all_numeric_predictors())



player_knn_tune <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) |>
                set_engine("kknn") |>
                set_mode("classification")
k_vals <- tibble(neighbors = seq(from = 1, to = 14, by = 2))

players_results <- workflow() |>
                    add_recipe(player_recipe) |>
                    add_model(player_knn_tune) |>
                    tune_grid(resamples = player_vfold, grid = k_vals) |>
                    collect_metrics()

players_results
```

In this look at the data, we first divide it into two parts, training and test, with a 75-25 split based on the `subscribe` column, making sure both chunks reflect the same mix of outcomes. For a solid guess on how well the model works while building it, we use 5 rounds of cross-validation, also grouped by `subscribe`, keeping the ratio of classes steady across the dataset. Ahead of running the model, we set up steps to adjust numerical inputs by centering and scaling them. That matters since the technique uses distances between points; without adjusting, bigger-scale numbers could take over those distances. We pick a flexible KNN setup by leaving the neighbor count as `tune()` - this way, instead of guessing, we can test which `k` works best. Then, we set up a list of odd numbers from 1 to 14 for `k`, so no ties happen during predictions while checking different group sizes.

```r
accuracies <- players_results |>
            filter(.metric %in% "accuracy")

accuracies_versus_k <- ggplot(accuracies, aes(x = neighbors, y = mean)) + 
                    geom_point() + 
                    geom_line() + 
                    labs(x = "Neighbors", y = "Accuracy Estimate") +
                    ggtitle("Fig. 5: Best Accuracy by Number of Neighbours") +
                    theme(text = element_text(size = 12))


accuracies_versus_k

```

```r
best_k <- accuracies |>
            arrange(desc(mean)) |>
            head(1) |>
            pull(neighbors)
best_k
```

The accuracy-vs-k chart displays shifts in model results when adjusting neighbor count. As k grows, performance improves, climbing until hitting a maximum at `k = 11`. This suggests `k = 11` outperforms other values tried. Because the top score occurs here on the curve, we choose `k = 11` as the ideal setting for our KNN setup.


```r
player_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = best_k) |>
                set_engine("kknn") |>
                set_mode("classification")

players_workflow <- workflow() |>
                    add_recipe(player_recipe) |>
                    add_model(player_spec) |>
                    fit(data = players_train)

players_workflow
```

Using the optimal value of `k = 11`, we finalized the KNN model within our workflow. The fitted results confirm that this choice gives the lowest misclassification rate with the rectangular kernel.

```r
player_test_predictions <- predict(players_workflow, players_test) |>
bind_cols(players_test)

player_test_predictions |>
metrics(truth = subscribe, estimate = .pred_class) |>
filter(.metric == "accuracy")

player_test_predictions |>
precision(truth = subscribe, estimate = .pred_class)

player_test_predictions |>
recall(truth = subscribe, estimate = .pred_class)

confusion <- player_test_predictions |>
conf_mat(truth = subscribe, estimate = .pred_class)

confusion
```

```r
prediction_bar_plot <- ggplot(player_test_predictions, aes(x = subscribe, fill = .pred_class)) +
geom_bar(position = "fill") +
labs(title = "Fig. 6: Proportion of Predictions vs. Actual Subscription Status",
    x = "Actual Subscription Status",
    y = "Proportion",
    fill = "Predicted Class")

prediction_bar_plot
```

**Discussion**

The question we started with was: "What player characteristics and behaviours are most predictive of subscribing to a game-related newsletter, and how do these features differ between various player types?" We then specified and narrowed that question to a statistical question we could ask, which was: "Can players experience and total hours played predict whether a player subscribes to the newsletter using the *players* dataset?" We found that we were able to build a predictive model with 55% accuracy to determine if a player would subscribe to the gaming newsletter based on their experience and hours. The model most accurately predicted "TRUE" to if someone was subscribed correctly, with a 47% success rate, but it only successfully predicted "FALSE" 8% of the time. "TRUE" was miscategorized as "FALSE" 26% of the time, and "FALSE" was miscategorized as "TRUE" 18% of the time. Therefore, this model is not super reliable. However, if we are to assess the data visualizations, we can see that hours of play is a slightly better predictor of subscription status (Fig. 4) than experience level (Fig. 2) as the proportion of "TRUE" to "FALSE" in every experience level is extremely similar. While more hours of play does have a positive correlation to subscription status, it is weakly correlated. Ultimately, our findings were that neither experience level nor hours of play could predict subscription status to a gaming magazine with reasonable accuracy, at least not with the provided data.

Our collective intuition going into this project was that players with higher experience and hours would be more likely to subscribe to a gaming newsletter. Upon initial skimming of the *players* dataframe, we noticed that the "Pro" experience variable did seem to correlate slightly more often with "TRUE", but no strong correlations were visible just by looking. We also saw at a glance that almost all players with over 10 hours of playtime were subscribed, but many also had zero hours, weakening that correlation. So, compared to our expectations that there would be a slight correlation of newsletter subscription with high play hours and experience, we found that the correlation was even weaker than expected.

The impact of these findings is not direct, as they do not give the team running the gaming newsletter immediate answers as to what demographic to target.
